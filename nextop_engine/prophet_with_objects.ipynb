```python
import pandas as pd
```

```python
from datetime import datetime
```

```python
df= pd.ExcelFile('.\\_element\\data\\private\\KPP일별투입(10_17).xlsx')
```

```python
df.sheet_names
```

```python
df_txs= pd.DataFrame(columns= ['구분','타입','코드','발송일','유형','수량'])
```

```python
type(df_txs)
```

```python
for sheet_name in df.sheet_names:
    df_sheet= df.parse(sheet_name)
    df_txs= pd.concat([df_txs, df_sheet])
```

```python
df_txs.info()
```

```python
date_change= lambda x: datetime(int(x[:4]),int(x[4:6]),int(x[6:])) if len(x)==8 \
                            else datetime.strptime(x, "%Y-%m-%d")
```

```python
df= df_txs
```

```python
df['발송일']=df['발송일'].map(str).map(date_change)
df= df.groupby([df['발송일'], df['유형']])['수량'].sum().unstack('유형')
df.fillna(0, inplace= True)
df['y_sum']= df.sum(axis=1)
```
